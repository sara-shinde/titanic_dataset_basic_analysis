Now we compute the survival rates for males and females using yet another approach. Which might fail.

```python
def compute_survival_rate(df, column):
    passenger_ct = df[column].count()
    survival_ct = df[column].sum()
    return survival_ct / passenger_ct
```

```python
compute_survival_rate(dataset, "Survived")
```

We have tested that the function works as intended.

The number of men in the second and third classes is large enough that these rates cannot be brushed aside as coincidence. Perhaps there is a good explanation about why these survival rates are the way they are.

Another interesting point is that the survival rate for men in the first class is lower than that of women in the third class.
